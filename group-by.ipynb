```python
# Company XYZ sales information
# Dictionry containing needed data
data = {'Sales Person':'Sam Charlie Amy Vanessa Carl Saray'.split(), 
        'Product':'Hp Hp Apple Apple Dell Dell '.split(),
        'Sales':[200,120,340,124,243,350]}
print('XYZ sales information\n_________________________') # print info.

serial = list(range(1,7))                       # row names from 1-6
df = pd.DataFrame(data,serial)                  # build data frame
df
```

```python
# finding sales information by product
print('Total items sold: by product')
df.groupby('Product').sum()
```

```python
df.groupby('Product').count()
```

```python
# Getting better info using describe()
df.groupby('Product').describe()
```

```python
# Defining a dictionary of values
d1 = {'A':'A1 A2 A3'.split(),'B':'B1 B2 B3'.split(),
      'C':'C1 C2 C3'.split()}
d2 = {'A':'A4 A5 A6'.split(),'B':'B4 B5 B7'.split(),
      'C':'C4 C5 C6'.split()}

# Now the data frames

df1 = pd.DataFrame(d1, index = '1 2 3'.split())
df2 = pd.DataFrame(d2,index = '4 5 6'.split())

# Concatenating
print('Row-wise concatenation:\n')
pd.concat([df1,df2]) # row-wise axis = 0
print('Column-wise concatenation:\n')
pd.concat([df1,df2],axis=1, sort=True)
```

```python
df2.index = '1 2 3'.split() # setting same index as df1

# This should then work
pd.concat([df1,df2],axis=1) # column-wise axis = 1
```

```python
df2.columns = 'D E F'.split()
pd.concat([df1,df2],axis=1)
```
