```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
%matplotlib inline
```

```python
# create plot values
x = np.linspace(0,10)                 # x-axis/time-scale
y = np.cos(x)    
```

```python
# functional plot
plt.plot(x,y)
```

```python
np.cos(0)
```

```python
np.cos(9)
```

```python
z = np.sin(x) # Adding an extra plot variable
plt.plot(x,y,x,z); plt.show()
```

```python
plt.plot(x,y,x,z)
plt.xlabel('Time axis')              # labeling x-axis
plt.ylabel('Magnitude')              # labeling y-axis
plt.title('Sine and Cosine wawes')   # graph title
plt.show()                           # printing
```

```python
plt.plot(x,y,x,z)
plt.xlabel('Time axis')
plt.ylabel('Magnitude')
plt.title('Sine and Cosine wawes')
plt.legend(['y','z'],loc=0)                  # loc=0 means best location

plt.show()
```

```python
# subploting
plt.subplot(1,2,1)                        # plot 1
plt.plot(x,y)
plt.title('Cosine plot')

plt.subplot(1,2,2)                        # plot 2
plt.plot(x,z)
plt.title('Sine plot')

plt.tight_layout()                        # avoid overall ping plots
plt.show()
```

```python
plt.subplot(1,2,1)                         # plot 1
plt.plot(x,y, 'r-x')                       # red plot with -x marker
plt.title('Cosine plot')

plt.subplot(1,2,2)                         # plot 2
plt.plot(x,z, 'g-o')                       # green plot
plt.title('Sine plot')

plt.tight_layout()                         # still avoiding overallping plots
plt.show()
```

```python
x = np.arange(0,10,2)
y = x**2
z = 2*x**2 + 3*x
plt.plot(x,y,'g-o')
plt.plot(x,z,'b-o')
plt.xlabel('Time axis')
plt.ylabel('Magnitude')
plt.title('PLots of x')
plt.legend(['y = x^2','z = 2x^2 + 3x'],loc=2)

plt.show()
```
